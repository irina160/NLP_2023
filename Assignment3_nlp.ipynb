Formally, given the document $d$, consisting of words $\left[ v_1, v_2, ..., v_{|d|} \right]$, the document representation $\mathbf{e}_d$ is defined as:

```python
#LOAD GLOVE VECTORS (funcction)

import numpy as np

def load_glove_vectors(file_path):
    word_vectors = {}
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            values = line.split()
            word = values[0]
            vector = np.array(values[1:], dtype='float32')
            word_vectors[word] = vector
    return word_vectors

```

```python
# reuse the code from Assignment 1-----------------------------------------------------------------------------------

import os
import glob
import re
import itertools
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import MWETokenizer
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.corpus import stopwords
from collections import Counter

def split_into_texts_and_labels(data):
    labels = []
    texts = data.split('\n')
    texts = [text for text in texts if len(text)>0]
    for i, text in enumerate(texts):
        try:
            nr = int(text[-1])
            labels.append(text[-1])
            texts[i] = ''.join(list(text)[:-2])
        except:
            break
    return texts, labels

def perform_basic_preprocessing(text):
    text = text.lower()

    for d in re.finditer('\d{1,2}[/-]\d{1,2}[/-]\d{2,4}', text):
        text = text.replace(str(d.group()), '<date>')

    for d in re.finditer(r'(\d{1,2}[\s]?|\d{1,2}[\s]?th[\s]?)?(?:january|february|march|april|may|june|july|august|september|october|november|december|jan|feb|mar|apr|may|jun|jul|aug|sep|oct|nov|dec)[\s-]?(\d{1,2})?[,\s-]?[\s]?\d{4}', text, re.I|re.M):
        text = text.replace(str(d.group()), '<date>')

    for d in re.finditer(r'(\d{1,2}|\d{1,2}[\s]?th)[\s]?(?:january|february|march|april|may|june|july|august|september|october|november|december|jan|feb|mar|apr|may|jun|jul|aug|sep|oct|nov|dec)', text, re.I|re.M):  
        text = text.replace(str(d.group()), '<date>')

    for d in re.finditer(r'\d{1,2}[/-]\d{4}', text, re.M|re.I):
        text = text.replace(str(d.group()), '<date>')
        
    text = re.sub('[0-9]*[[,.]?[0-9]+]*', '<num>', text)
    
    text = re.sub('([^a-z<>])+', ' ', text)

    return text

def perform_tokenization(text):
    stop_words = set(stopwords.words('english'))
    lemmatizer = WordNetLemmatizer()

    tokenizer = MWETokenizer([('<', 'num', '>'), ('<', 'date', '>')])
    tokens = tokenizer.tokenize(nltk.word_tokenize(text))
    
    tokens = [token for token in tokens if not token in stop_words]
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return tokens
```

```python
# Document embedding as the average of word embeddings--------------------------------------------------------------

def get_doc_repr(doc, word_embeddings):
    tokens = perform_tokenization(perform_basic_preprocessing(doc))
    # remove Nones, does happen if the token is not in the dictionary, just ignore the token
    tokens = [token for token in tokens if token in smaller_word_dict]
    # get the embeddings of the tokens and calculate an average embedding
    e_d = np.sum(list(map(word_embeddings.get, tokens)), axis=0) / len(tokens) 
    return e_d
```

```python
# Define method for batching
def make_batches(dataset, batch_size, max_doc_len):
    '''
    Each batch is a two-dimensional matrix of batch-size to max-document-length, 
    containing the IDs of the words in the corresponding documents. 
    If a document has more than max-document-length words, 
    only the first max-document-length words should be kept.
    '''
    #batch = np.zeros((batch_size, max_doc_len))
    batches = []
    if batch_size <2: raise ValueError("batch size has to be at least 2")
    if max_doc_len <50: raise ValueError("max_doc_len has to be at least 50")

    for doc in dataset:
        #doc_vector = [word_to_id[word] for word in doc[:max_doc_len]]
        doc_vector = [id for id in doc[:max_doc_len]]
        # Pad the vector with zeros if the document is shorter than max_doc_len
        doc_vector += [0] * (max_doc_len - len(doc_vector))
        batches.append(doc_vector)

    # Split into batches
    batches = [batches[i:i + batch_size] for i in range(0, len(batches), batch_size)]

    return batches
```

```python
# Create word embedding lookup 
import torch
from torch.nn import Embedding

def create_emb_lookup(embedding_dict, word_dic):
    '''
    Uses the pre-trained vectors of a word embedding model (like word2vec or GloVe) 
    to initialize the word embeddings of the lookup. The lookup is a matrix which maps the ID of each word to the corresponding word vector. 
    If the vector of a word in the lookup does not exist 
    in the pretrained word embeddings, the corresponding vector is initialized randomly.
    '''
    #some_embedding = next(iter(embedding_dict.values()), None) #this is just to calc the required lenth of embedding
    lookup = []
    #emb_dim = (len(word_dic.items()), len(some_embedding))
    
    for word, word_id in word_dic.items():
        if word in embedding_dict:
            lookup.append(embedding_dict[word]) #these will be the
        else:
            lookup.append(np.random.rand(300))

    lookup = torch.tensor(lookup, dtype=torch.float64)
    emb = Embedding.from_pretrained(lookup)

    # return the lookup as a torch.nn.Embedding
    return emb 
```

```python
# Define the model which receives the word embedding lookup, and includes it in the model as model's parameters.
import torch
import torch.nn as nn
import torch.nn.functional as F
from tqdm import tqdm

class ClassificationAverageModel(nn.Module):
    def __init__(self, embedding_lookup, num_classes):
        super(ClassificationAverageModel, self).__init__()

        # word Embedding Layer
        self.embedding = embedding_lookup

        # Other model parameters??
        # ...

        # output Layer
        self.fc = nn.Linear(embedding_lookup.embedding_dim, num_classes)

    def forward(self, input_ids):
        """
        Forward pass of the model.
        
        Parameters:
        - input_ids: Tensor of shape (batch_size, max_doc_len)
        
        Returns:
        - probabilities: Tensor of shape (batch_size, num_classes)

         document embedding is utilized to predict the probability of the output classes, 
         done by applying a linear transformation from the embeddings size to the number of classes, 
         followed by Softmax. The linear transformation also belongs to the model's parameters 
         and will be learned in training.
        """
        # Perform embedding lookup
        embedded = self.embedding(input_ids)
    
        # Mask for padding
        mask = (input_ids != 0).unsqueeze(-1).float()  # assume that padding ID is 0
    
        # Calculate the element-wise mean along the sequence dimension
        mean_embedding = torch.sum(embedded * mask, dim=1) / torch.sum(mask, dim=1)
    
        # Forward pass through the output layer
        logits = self.fc(mean_embedding)
    
        # Apply softmax to get probabilities
        probabilities = F.softmax(logits, dim=-1)
    
        return probabilities
```
